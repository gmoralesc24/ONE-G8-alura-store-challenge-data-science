```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# URLs de los datasets
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargar los datasets
tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Unir todos los datasets en uno solo para facilitar el análisis
# Es importante asegurarse de que todas las columnas tienen el mismo nombre y tipo de dato
# Antes de unirlos, vamos a limpiar los nombres de las columnas para asegurar consistencia
def limpiar_nombres_columnas(df):
    df.columns = df.columns.str.strip().str.replace(' ', '_').str.replace('del_', '').str.lower()
    return df

tienda = limpiar_nombres_columnas(tienda)
tienda2 = limpiar_nombres_columnas(tienda2)
tienda3 = limpiar_nombres_columnas(tienda3)
tienda4 = limpiar_nombres_columnas(tienda4)

# Unir los dataframes
tiendas_unidas = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

# Convertir la columna 'fecha_de_compra' a formato datetime
tiendas_unidas['fecha_de_compra'] = pd.to_datetime(tiendas_unidas['fecha_de_compra'], format='%d/%m/%Y')

print("Primeras 5 filas del dataset unido:")
print(tiendas_unidas.head())
print("\nInformación general del dataset unido:")
print(tiendas_unidas.info())
print("\nEstadísticas descriptivas del dataset unido:")
print(tiendas_unidas.describe())

# 1. Análisis de facturación
# Calcular el total facturado por la tienda (Precio + Costo de envío)
tiendas_unidas['total_facturado'] = tiendas_unidas['precio'] + tiendas_unidas['costo_de_envío']

# Facturación total
facturacion_total = tiendas_unidas['total_facturado'].sum()
print(f"\n1. Análisis de facturación:")
print(f"La facturación total de Alura Store es: ${facturacion_total:,.2f}")

# Facturación por mes/año (para ver tendencias)
tiendas_unidas['año_mes'] = tiendas_unidas['fecha_de_compra'].dt.to_period('M')
facturacion_por_mes = tiendas_unidas.groupby('año_mes')['total_facturado'].sum().reset_index()
facturacion_por_mes['año_mes'] = facturacion_por_mes['año_mes'].astype(str) # Convertir a string para graficar

plt.figure(figsize=(12, 6))
sns.lineplot(x='año_mes', y='total_facturado', data=facturacion_por_mes)
plt.title('Facturación Mensual de Alura Store')
plt.xlabel('Año-Mes')
plt.ylabel('Facturación Total ($)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# 2. Ventas por categoría
# Contar el número de ventas por cada categoría de producto
ventas_por_categoria = tiendas_unidas['categoría_producto'].value_counts().reset_index()
ventas_por_categoria.columns = ['categoría_producto', 'cantidad_ventas']

print(f"\n2. Ventas por categoría:")
print("Cantidad de ventas por categoría de producto:")
print(ventas_por_categoria)

# Visualización de ventas por categoría
plt.figure(figsize=(10, 7))
sns.barplot(x='cantidad_ventas', y='categoría_producto', data=ventas_por_categoria, palette='viridis')
plt.title('Cantidad de Ventas por Categoría de Producto')
plt.xlabel('Cantidad de Ventas')
plt.ylabel('Categoría del Producto')
plt.tight_layout()
plt.show()

# También podemos ver la facturación por categoría
facturacion_por_categoria = tiendas_unidas.groupby('categoría_producto')['total_facturado'].sum().sort_values(ascending=False).reset_index()
print("\nFacturación total por categoría de producto:")
print(facturacion_por_categoria)

plt.figure(figsize=(10, 7))
sns.barplot(x='total_facturado', y='categoría_producto', data=facturacion_por_categoria, palette='magma')
plt.title('Facturación Total por Categoría de Producto')
plt.xlabel('Facturación Total ($)')
plt.ylabel('Categoría del Producto')
plt.tight_layout()
plt.show()

# 3. Calificación promedio de la tienda
# Calcular la calificación promedio de todas las ventas
calificacion_promedio_global = tiendas_unidas['calificación'].mean()
print(f"\n3. Calificación promedio de la tienda:")
print(f"La calificación promedio global de Alura Store es: {calificacion_promedio_global:.2f}")

# Calificación promedio por vendedor
calificacion_por_vendedor = tiendas_unidas.groupby('vendedor')['calificación'].mean().sort_values(ascending=False).reset_index()
print("\nCalificación promedio por vendedor:")
print(calificacion_por_vendedor)

plt.figure(figsize=(10, 6))
sns.barplot(x='calificación', y='vendedor', data=calificacion_por_vendedor, palette='coolwarm')
plt.title('Calificación Promedio por Vendedor')
plt.xlabel('Calificación Promedio')
plt.ylabel('Vendedor')
plt.tight_layout()
plt.show()

# 4. Productos más y menos vendidos
# Contar la ocurrencia de cada producto (ventas de unidades)
productos_vendidos = tiendas_unidas['producto'].value_counts()

# Productos más vendidos (Top 10)
productos_mas_vendidos = productos_vendidos.head(10).reset_index()
productos_mas_vendidos.columns = ['producto', 'cantidad_vendida']
print(f"\n4. Productos más y menos vendidos:")
print("Top 10 Productos más vendidos:")
print(productos_mas_vendidos)

plt.figure(figsize=(12, 7))
sns.barplot(x='cantidad_vendida', y='producto', data=productos_mas_vendidos, palette='rocket')
plt.title('Top 10 Productos Más Vendidos')
plt.xlabel('Cantidad Vendida')
plt.ylabel('Producto')
plt.tight_layout()
plt.show()

# Productos menos vendidos (Bottom 10)
productos_menos_vendidos = productos_vendidos.tail(10).reset_index()
productos_menos_vendidos.columns = ['producto', 'cantidad_vendida']
print("\nTop 10 Productos menos vendidos:")
print(productos_menos_vendidos)

plt.figure(figsize=(12, 7))
sns.barplot(x='cantidad_vendida', y='producto', data=productos_menos_vendidos, palette='mako')
plt.title('Top 10 Productos Menos Vendidos')
plt.xlabel('Cantidad Vendida')
plt.ylabel('Producto')
plt.tight_layout()
plt.show()

# 5. Envío promedio por tienda (Lugar de Compra)
# Calcular el costo de envío promedio por 'Lugar de Compra'
envio_promedio_por_lugar = tiendas_unidas.groupby('lugar_de_compra')['costo_de_envío'].mean().sort_values(ascending=False).reset_index()
print(f"\n5. Envío promedio por tienda (Lugar de Compra):")
print("Costo de envío promedio por lugar de compra:")
print(envio_promedio_por_lugar)

plt.figure(figsize=(12, 7))
sns.barplot(x='costo_de_envío', y='lugar_de_compra', data=envio_promedio_por_lugar, palette='flare')
plt.title('Costo de Envío Promedio por Lugar de Compra')
plt.xlabel('Costo de Envío Promedio ($)')
plt.ylabel('Lugar de Compra')
plt.tight_layout()
plt.show()
```
